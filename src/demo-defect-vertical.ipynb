```python
# Librosa.
import librosa
import librosa.display

# Plotting and numeric libraries.
import IPython.display as ipd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_plt(data, sr):
    """
    Show plot.
    :param data: data for plotting
    :param sr: sample rate
    :param y_axis: text for y_axis
    :param title: plot title
    """

    # Create figure and plot graphs onto it.
    plt.figure(figsize=(20, 5))
    librosa.display.specshow(data, sr=sr, x_axis='time', y_axis='hz', cmap='turbo')
    plt.colorbar(format='%+02.0f dB')
```

```python
# Get spectre fun.
get_spectre = lambda d: librosa.amplitude_to_db(abs(librosa.stft(d)))

# Cur data into two parts.
cut_data = lambda d: (d[:len(d) // 2], d[len(d) // 2:])
```

```python
# Analyze 0001 wav.
y, sr = librosa.load('../wavs/origin/0057.wav', mono=False, sr=None)
```

```python
# Original spectre.
s = get_spectre(y[0])
print(s.shape)
# shape is (hz, time)
show_plt(s, sr)
print(s.shape)
```

```python
# Cut and show spectre.
y1, y2 = cut_data(y[0])
s1, s2 = get_spectre(y1), get_spectre(y2)
print(s1.shape, s2.shape)
show_plt(s1, sr)
show_plt(s2, sr)
print(s.shape)
```

```python
# Get spectre and then cut it.
s1, s2 = cut_data(get_spectre(y[0]).T)
print(s1.shape, s2.shape)
show_plt(s1.T, sr)
show_plt(s2.T, sr)
print(s.shape)
```

```python
# Experiments.
print(s.shape)
t = s.T
# 0001.wav - 541:545
# 0002.wav - 154:158
# 0013.wav - 40:50
t = t[163:167]
new_s = t.T
show_plt(new_s, sr)
```
